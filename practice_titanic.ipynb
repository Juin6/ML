피클래스 : 1등석은 상대적으로 생존률이 높을 것이다.
성별 : 여성의 생존률이 더 높을 것이다.
나이 : 어린아이는 생존률이 높을 것이다. 노인은 생존률이 높을수도 있고 낮을 수도 있다.
가족여부 : 자녀 없이 가족이 있을 경우 생존률이 낮을 것이다. 함께 탈려고 하기 때문에.
자녀여부 : 자녀가 있을 경우 생존률이 높을 것이다.
티켓 : 일행인지를 유추할 수 있을 것이다. 그러나 직접적으로 분석하기에는 어려움이 있을 것이다.
요금 : 요금이 높을 수록 경제력이 높을 것이기에 생존확률이 높을 것이다.
탑승위치 : 탑승위치는 프랑스, 아일랜드, 영국이 있다. 언어가 다르므로 재난상황에서 커뮤니케이션 능력에 따라 생존여부에 영향을 미쳤을 것이다.

```python
def my_data(x):#x는train_data/test_data
    x['Cabin'] = x['Cabin'].isnull()
    features = ['Pclass', 'Sex', 'Cabin', 'SibSp']
    X = pd.get_dummies(x[features])
    return X
train_data1 = my_data(train_data)
test_data1 = my_data(test_data)
```
